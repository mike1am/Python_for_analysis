```

Отсортируйте от меньшей стоимости к большей

3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
```

````python
import re

def numFormat(col):
    return col.apply(
        lambda val:
            f"{val:_.2f}".replace("_", " ")
            if re.fullmatch(r"((\-\d+)|(\d*))(\.\d+)?", str(val))
            else val
    )
````

### 3.1

````python
df.groupby('bedrooms') \
    .agg({'price': ['mean']}) \
    .sort_values(by=('price', 'mean')) \
    .apply(numFormat) # при указании списка в {'price': ['mean']} в выходной форме выводится операция
````

### 3.2

````python
df.pivot_table(
    index='condition',
    values='price',
    aggfunc=['min', 'mean', 'max']
).apply(numFormat)
````

### 3.3

````python
pd.crosstab(
    index=df['waterfront'],
    columns=df['view']
)
````

### 3.4

````python
ct = pd.crosstab(
    index=df['floors'],
    columns=df['bedrooms'],
    margins=True
) \
.sort_values(by='All', ascending=False) \
.sort_values(by='All', ascending=False, axis=1)
ct
````

Больше всего одноэтажных домов (10680). Также больше всего домов в 3-мя спальнями (9824).

Больше всего домов из сочетаний этажи-спальни:

````python
maxVar = ct.iloc[1:, 1:].max().max()

f"Этажей: {ct.index[ct.eq(maxVar).any(axis=1)][0]}, " + \
f"спален: {ct.columns[ct.eq(maxVar).any(axis=0)][0]}, " + \
f"домов: {maxVar}"
````

### 3.5

````python
df.pivot_table(
    index='grade',
    columns='condition',
    values='price',
    aggfunc='median',
    fill_value="---"
).apply(numFormat)
````
